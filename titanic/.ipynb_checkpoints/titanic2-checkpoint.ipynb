```python
# turn 'Sex' into int
train_df['Sex'] = train_df['Sex'].map( {'female': -1, 'male': 1} ).astype(int)
test_df['Sex'] = test_df['Sex'].map( {'female': -1, 'male': 1} ).astype(int)
```

```python
# simplify the cabin
def cabin_class(x):
    return str(x)[0]
        
train_df['Cabin'] = train_df['Cabin'].apply(cabin_class)
test_df['Cabin'] = test_df['Cabin'].apply(cabin_class)
```

```python
train_df[['Cabin', 'Survived']].groupby(['Cabin'], as_index=False).agg(['mean','count'])
```

```python
# turn 'Cabin' into float
cabin_map = {'A':0.47,'B':0.745,'C':0.59,'D':0.758,'E':0.750,'F':0.615,'G':0.5,'n':0.3,'T':0}

train_df['Cabin'] = train_df['Cabin'].map(cabin_map).astype(float)
test_df['Cabin'] = test_df['Cabin'].map(cabin_map).astype(float)
```

```python
train_df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).agg(['count','mean'])
```

```python
def embark_map(x):
    dic = {'C': 0.55, 'Q': 0.39, 'S': 0.34}
    if x in dic:
        return dic[x]
    else:
        return train_df['Survived'].mean()

train_df['Embarked'] = train_df['Embarked'].apply(embark_map)
test_df['Embarked'] = test_df['Embarked'].apply(embark_map)
train_df.head()
```


### Fill the missing 'Age'
From the correlatiom matrix, we found 'Age' is highly correlated to 'Pclass' and 'SibSp', so we decide to used these datasets to predict missing 'Age' values.  

We don't have so many data availabe for 'SibSp' = 2, 3, 4..., so we treat them as a category.  
For 'SibSp' = `[0,1]` and 'Pclass' = `[1,2,3]`, we group them into six categories.   


```python
for dataset in total:
    
    fix_age = np.zeros((2, 3))
    
    for i in range(0, 2):
        for j in range(1, 4):
            
            part = dataset[(dataset['SibSp'] == i) & \
                                      (dataset['Pclass'] == j)]['Age'].dropna()
            fix_age[i, j-1] = part.mean() 
    
    for i in range(len(dataset)):
        
        if math.isnan(dataset['Age'].iloc[i]):
            
            if dataset['SibSp'].iloc[i] >= 2:
                
                dataset.at[i, 'Age'] = dataset['Age'][(dataset['SibSp'] >= 2)].dropna().mean()
                
            for j in range(0, 2):
                for k in range(1, 4):
                    if dataset['SibSp'].iloc[i] == j and dataset['Pclass'].iloc[i] == k:
                        
                        dataset.at[i, 'Age'] = fix_age[j, k-1]
            
#train_df.describe(include='all')
#test_df.describe(include='all')
train_df.corr()
```

```python
title = ['NA'] * length1
prefix = ['Mr', 'Mrs', 'Miss']
for i in range(length1):
    for j in range(len(prefix)):
        
        if str(train_df['Name'].iloc[i]).find(prefix[j]) != -1:
            
            title[i] = prefix[j]
            
title2 = ['NA'] * length2
prefix = ['Mr', 'Mrs', 'Miss']
for i in range(length2):
    for j in range(len(prefix)):
        
        if str(test_df['Name'].iloc[i]).find(prefix[j]) != -1:
            
            title2[i] = prefix[j]
            
train_df['Title'] = title
test_df['Title'] = title2
train_df[['Title','Survived']].groupby(['Title'], as_index=False).agg(['count','mean','std'])
```

Now, we are going to make the mnumerical.


```python
title_map = {'Mr': 0.16, 'Mrs': 0.8, 'Miss': 0.7, 'NA': 0.5}

train_df['Title'] = train_df['Title'].map(title_map).astype(float)
test_df['Title'] = test_df['Title'].map(title_map).astype(float)

train_df.corr()
```


### Prepare dataset for machine learning

Drop 'PassengerId', 'Name', and 'Ticket'.  


```python
#X_train = train_df.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'SibSp', 'Parch'], axis=1)
X_train = train_df.drop(['PassengerId', 'Survived', 'Name', 'Ticket'], axis=1)
Y_train = train_df["Survived"]

# X_test  = test_df.drop(['PassengerId', 'Name', 'Ticket','SibSp', 'Parch'], axis=1)
X_test  = test_df.drop(['PassengerId', 'Name', 'Ticket'], axis=1)

X_train.shape, Y_train.shape, X_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.svm import SVC, LinearSVC

train_set = 600

# clf = LinearSVC(C=0.3)
# clf = SVC(kernel="rbf", gamma=5, C=0.001)
clf = SVC(kernel="poly", degree=3, coef0=1, C=5)

clf.fit(X_train[:train_set], Y_train[:train_set])
Y_pred = clf.predict(X_test)

acc_tr = round(clf.score(X_train[:train_set], Y_train[:train_set]), 4)
acc_va = round(clf.score(X_train[train_set+1:], Y_train[train_set+1:]), 4)

print('Accuracy of traing set: ', acc_tr)
print('Accuracy by cross validation set: ', acc_va)
```

```python
from sklearn.tree import DecisionTreeClassifier

train_set = 600

# clf = Pipeline((
#     ("scaler", StandardScaler()),
#     ("linear_svc", LinearSVC(C=0.5, loss="hinge")),
# ))

clf = DecisionTreeClassifier(max_depth=4)

clf.fit(X_train[:train_set], Y_train[:train_set])
Y_pred = clf.predict(X_test)

acc_tr = round(clf.score(X_train[:train_set], Y_train[:train_set]), 4)
acc_va = round(clf.score(X_train[train_set+1:], Y_train[train_set+1:]), 4)

print('Accuracy of traing set: ', acc_tr)
print('Accuracy by cross validation set: ', acc_va)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestClassifier

# clf = Pipeline((
#     ("scaler", StandardScaler()),
#     ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
# ))
# clf = SVC(kernel="rbf", gamma=5, C=0.0001)
clf = RandomForestClassifier(n_estimators=600, max_leaf_nodes=20, n_jobs=-1)

clf.fit(X_train[:train_set], Y_train[:train_set])
Y_pred = clf.predict(X_test)

acc_tr = round(clf.score(X_train[:train_set], Y_train[:train_set]), 4)
acc_va = round(clf.score(X_train[train_set+1:], Y_train[train_set+1:]), 4)

print('Accuracy of traing set: ', acc_tr)
print('Accuracy by cross validation set: ', acc_va)
```
